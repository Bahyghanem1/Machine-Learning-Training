### 2. Rainfall Statistics
Design a program that lets the user enter the total rainfall for each of 12 months into a
list. The program should calculate and display the total rainfall for the year, the average
monthly rainfall, and the months with the highest and lowest amounts.

```python
Rain_Stats = []
print("Enter the number of rainfalls for every months in order starting from January")
for n in range(12):
  Total_Rain = int(input())
  Rain_Stats.append(Total_Rain)

print("The number of rainfalls for the year is: ", sum(Rain_Stats))
print("The average monthly rainfall: ", sum(Rain_Stats)/12)

print("The months with the highest rainfall:", max(Rain_Stats))
print("The months with the lowest rainfall:", min(Rain_Stats))
```

### 3. Larger Than n
In a program, write a function that accepts two arguments: a list, and a number n. Assume
that the list contains numbers. The function should display all of the numbers in the list
that are greater than the number n.

```python
def greater_than(List, n):
  for L in List:
    if L > n:
      print(L, end=" ")
def main():
  List = [2, 9, 4, 7]
  greater_than(List, 3)

main()
```

### 6. Alphabetic Telephone Number Translator
Many companies use telephone numbers like 555-GET-FOOD so the number is easier for
their customers to remember. On a standard telephone, the alphabetic letters are mapped
to numbers in the following fashion: <br>

A, B, and C = 2 <br>
D, E, and F = 3 <br>
G, H, and I = 4 <br>
J, K, and L = 5 <br>
M, N, and O = 6 <br>
P, Q, R, and S = 7 <br>
T, U, and V = 8 <br>
W, X, Y, and Z = 9 <br>
Write a program that asks the user to enter a 10-character telephone number in the format XXX-XXX-XXXX. The application should display the telephone number with any
alphabetic characters that appeared in the original translated to their numeric equivalent. For example, if the user enters 555-GET-FOOD the application should display
555-438-3663.

```python
tele_std = {"A": 2, "B": 2, "C": 2, "D": 3, "E": 3, "F": 3,
            "G": 4,"H": 4, "I": 4, "J": 5, "K": 5, "L": 5, 
            "M": 6,"N": 6, "O": 6, "P": 7, "Q": 7, "R": 7, 
            "S": 7,"T": 8, "U": 8, "V": 8, "W": 9, "X": 9, 
            "Y": 9, "Z": 9}

tele_char = input("Please Enter a 10-character telephone number \
in the format \'XXX-XXX-XXXX\': ")

num = ""
for i in tele_char:
    if i in tele_std:
        num += str(tele_std[i])
    else:
        num += i

print("The numeric equivalent is:", num)
```

### 7. Sentence Capitalizer
Write a program with a function that accepts a string as an argument and returns a copy of
the string with the first character of each sentence capitalized. For instance, if the argument
is “hello. my name is Joe. what is your name?” the function should return the string “Hello.
My name is Joe. What is your name?” The program should let the user enter a string and
then pass it to the function. The modified string should be displayed.

```python
def capital_string(s):
  lis = s.split('. ')
  print(lis)
  x = ''

  for it1 in range(len(lis)):
      x += lis[it1][0].capitalize()
      x +=lis[it1][1:]
      if it1 != (len(lis)-1):
        x += '. ' 
  print(x)

def main():
    s = input('Enter a sentence: ')
    capital_string(s) 

main()
```

### 8. Vowels and Consonants **(Extra)**
Write a program with a function that accepts a string as an argument and returns the
number of vowels that the string contains. The application should have another function that accepts a string as an argument and returns the number of consonants that the string
contains. The application should let the user enter a string and should display the number
of vowels and the number of consonants it contains. 

```python
def calc_num_vowels(string):
    vowels = 'aeiou'
    num_vowels = 0
    for ch in string:
        if ch.lower() in vowels:
            num_vowels += 1
    return num_vowels
 
def calc_num_consonants(string):
    consonants = 'bcdfghjklmnpqrstvwxyz'
    num_consonants = 0
    for ch in string:
        if ch.lower() in consonants:
            num_consonants += 1
    return num_consonants
 
users_string = input("Enter a string of characters: ")
 
vowels = calc_num_vowels(users_string)
consonants = calc_num_consonants(users_string)
 
print("Your string has:", vowels, "vowels and", consonants, "consonants.")
```

### 9. Most Frequent Character **(Extra)**


Write a program that lets the user enter a string and displays the character that appears
most frequently in the string

```python
def main():
    string = input('Enter a sentence: ')
    strings=string.lower()
    counter = 0
    total_counter = 0
    most_freq_char = ""
    for ch in strings:
        for str in strings:
            if str == ch and ch != ' ':  #not counting spaces
                counter += 1
        if counter > total_counter:
            total_counter = counter
            most_freq_char = ch
        counter = 0
    print("The most frequent character is '", most_freq_char, "' and it appears", total_counter, "times.")

main()
```

### 10. Word Separator **(Extra)**
Write a program that accepts as input a sentence in which all of the words are run together
but the first character of each word is uppercase. Convert the sentence to a string in which
the words are separated by spaces and only the first word starts with an uppercase letter.
For example the string “StopAndSmellTheRoses.” would be converted to “Stop and smell
the roses.”

```python
sentence = input("Enter your sentence: ")
new_sentence = sentence[0]
for ch in sentence[1:]:
    if ch.isupper():
        new_sentence += ' ' + ch.lower()
    else:
        new_sentence += ch
print(new_sentence)
```

### 11. Pig Latin **(Extra)**
Write a program that accepts a sentence as input and converts each word to “Pig Latin.” In
one version, to convert a word to Pig Latin you remove the first letter and place that letter
at the end of the word. Then you append the string “ay” to the word. Here is an example: <br>
English:        **I SLEPT MOST OF THE NIGHT** <br>
Pig Latin:**IAY LEPTSAY OSTMAY FOAY HETAY IGHTNAY**<br>

```python
sentence = input('Enter your sentence: ')
words = sentence.split(' ')
new_words = []
for word in words:
    if word[-1].isalpha():
        new_words.append(word[1:] + word[0] + 'AY')
    else:
        new_words.append(word[1:-1] + word[0] + 'AY' + word[-1])      
 
new_sentence = ' '.join(new_words)
 
print(new_sentence)
```
